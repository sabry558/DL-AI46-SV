```python
import pandas as pd
import torch
```

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist.data, mnist.target

print(f"Data shape: {X.shape}")
print(f"Labels shape: {y.shape}")
```

### get subset of the data to check sanity

```python
x_subset = X[:50]
y_subset = y[:50]
x_tensor = torch.tensor(x_subset, dtype=torch.float32)
y_tensor = torch.tensor(y_subset.astype(int), dtype=torch.long)
```

```python
x_tensor.shape
```

```python
class NN(torch.nn.Module):
    def __init__(self):
        super(NN, self).__init__()
        #define the layers of the neural network

        #first layer
        self.fc1 = torch.nn.Linear(784, 128)

        #second layer
        self.fc2 = torch.nn.Linear(128, 256)

        #third layer
        self.fc3=torch.nn.Linear(256, 128)

        #output layer
        self.fc4 = torch.nn.Linear(128, 18)
    def forward(self, x):
        #apply activation function to the output of each layer except the last one
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = self.fc4(x)
        return x
#define the loss function which is cross entropy because our problem is multiclass classification    
criterion=torch.nn.CrossEntropyLoss()

# instantiate the model
model=NN()

# Adam optimizer updates model parameters using gradient
optimizer=torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
epoch=10
for epoch in range(epoch):
    #forward pass
    outputs=model(x_tensor)
    loss=criterion(outputs, y_tensor)

    #backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f"Epoch [{epoch+1}/{epoch}], Loss: {loss.item():.4f}")
    
```

### loss goes near zero so sanity check is passed

```python
# spliting the data into training and testing sets and convert it to tensors
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y, random_state=42)
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.astype(int), dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.astype(int), dtype=torch.long)
```

### define simple nn with only two layers


```python
class simpleNN(torch.nn.Module):
    def __init__(self):
        super(simpleNN, self).__init__()
        #define the layers of the neural network

        #first layer
        self.fc1 = torch.nn.Linear(784, 32)

        self.fc2 = torch.nn.Linear(32, 10)

        

    def forward(self, x):
        #apply activation function to the output of each layer except the last one
        x = torch.relu(self.fc1(x))
        x= self.fc2(x)
        return x
```

```python
epochs=20
model=simpleNN()
#define the loss
criterion=torch.nn.CrossEntropyLoss()
# define the optimizer
optimizer=torch.optim.Adam(model.parameters(), lr=0.001)

for epoch in range(epochs):
    #forward pass
    outputs=model(X_train_tensor)
    loss=criterion(outputs, y_train_tensor)

    #backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    #print the loss for each epoch
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
model.eval()
outputs=model(X_test_tensor)
_, predicted = torch.max(outputs.data, 1)
accuracy = (predicted == y_test_tensor).sum().item() / len(y_test_tensor)
print(f"Test Accuracy: {accuracy:.4f}")    
```

### increase the model complexity by adding more layers, wider hidden units

```python
class complexNN(torch.nn.Module):
    def __init__(self):
        super(complexNN, self).__init__()
        #define the layers of the neural network

        #first layer
        self.fc1 = torch.nn.Linear(784, 512)
        #second layer
        self.fc2 = torch.nn.Linear(512, 256)

        #third layer
        self.fc3=torch.nn.Linear(256, 128)
        #forth layer
        self.fc4=torch.nn.Linear(128,64)
        #output layer
        self.fc5=torch.nn.Linear(64, 10)



        

    def forward(self, x):
        #apply activation function to the output of each layer except the last one
        x = torch.relu(self.fc1(x))
        x=torch.relu(self.fc2(x))
        x=torch.relu(self.fc3(x))
        x= torch.relu(self.fc4(x))
        x=self.fc5(x)
        return x
```

```python
epochs=20
model=complexNN()
#define the loss
criterion=torch.nn.CrossEntropyLoss()
# define the optimizer
optimizer=torch.optim.Adam(model.parameters(), lr=0.001)

for epoch in range(epochs):
    #forward pass
    outputs=model(X_train_tensor)
    loss=criterion(outputs, y_train_tensor)

    #backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    #print the loss for each epoch
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
model.eval()
outputs=model(X_test_tensor)
_, predicted = torch.max(outputs.data, 1)
accuracy = (predicted == y_test_tensor).sum().item() / len(y_test_tensor)
print(f"Test Accuracy: {accuracy:.4f}")    
```

### same neural network like above with regularization (added dropout layer)


```python
class complexNN_withRegularization(torch.nn.Module):
    def __init__(self):
        super(complexNN_withRegularization, self).__init__()
        #define the layers of the neural network

        #first layer
        self.fc1 = torch.nn.Linear(784, 512)
        #second layer
        self.fc2 = torch.nn.Linear(512, 256)

        #third layer
        self.fc3=torch.nn.Linear(256, 128)
        #forth layer
        self.fc4=torch.nn.Linear(128,64)
        #output layer
        self.fc5=torch.nn.Linear(64, 10)

        #dropout layer with a dropout rate of 0.3
        self.dropout = torch.nn.Dropout(p=0.5)



        

    def forward(self, x):
        #apply activation function to the output of each layer except the last one
        x = torch.relu(self.fc1(x))
        x= self.dropout(x)
        x=torch.relu(self.fc2(x))
        x=self.dropout(x)
        x=torch.relu(self.fc3(x))
        x=self.dropout(x)
        x= torch.relu(self.fc4(x))
        x=self.dropout(x)
        x=self.fc5(x)
        return x
```

```python
epochs=20
model=complexNN()
#define the loss
criterion=torch.nn.CrossEntropyLoss()
# define the optimizer
optimizer=torch.optim.Adam(model.parameters(), lr=0.001)

for epoch in range(epochs):
    # we added model.trian() because we use dropout 
    model.train()
    #forward pass
    outputs=model(X_train_tensor)
    loss=criterion(outputs, y_train_tensor)

    #backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    #print the loss for each epoch
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
    
model.eval()
outputs=model(X_test_tensor)
_, predicted = torch.max(outputs.data, 1)
accuracy = (predicted == y_test_tensor).sum().item() / len(y_test_tensor)
print(f"Test Accuracy: {accuracy:.4f}")    
```
